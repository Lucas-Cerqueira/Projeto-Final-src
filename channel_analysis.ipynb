```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('two-ray-debug.csv', header=None, names=['tx_power', 'distance', 'channel_gain']).drop_duplicates()
df.head()
```

```python
df['db_m'] = 10 * np.log10(df['distance'])
```

```python
df = pd.read_csv('two-ray-rural-debug.csv', header=None, names=['tx_power', 'distance', 'channel_gain']).drop_duplicates()
df['db_m'] = 10 * np.log10(df['distance'])
ax = df.query('db_m <= 25 and db_m > 15')\
       .plot(x='db_m', y='channel_gain', grid=True, 
             xticks=[15, 20, 25],
             yticks=[-85, -90, -95, -100, -105])

df = pd.read_csv('two-ray-highway-debug.csv', header=None, names=['tx_power', 'distance', 'channel_gain']).drop_duplicates()
df['db_m'] = 10 * np.log10(df['distance'])
df.query('db_m <= 25 and db_m > 15')\
  .plot(x='db_m', y='channel_gain', grid=True, 
        xticks=[15, 20, 25],
        yticks=[-85, -90, -95, -100, -105],
        ax=ax)

ax.legend(['Rural', 'Highway'])
```

```python
df[['distance', 'channel_gain']].plot(x='distance', y='channel_gain', logx=True)
```

## Log distance - Urban

```python
df = pd.read_csv('log-urban-debug.csv', header=None, names=['tx_power', 'distance', 'channel_gain'])
df.head()
```

```python
df['db_m'] = 10 * np.log10(df['distance'])
```

```python
df.drop_duplicates(subset=['tx_power', 'distance'], keep='first')\
  .query('db_m < 25')\
  .plot(x='db_m', y='channel_gain')
```

```python
df.query[['db_m', 'channel_gain']].plot(x='db_m', y='channel_gain')
```
